#### 1. Percentile Calculation: 
Find the 95th percentile of earthquake magnitude in Japan where the magType is 'mb'.

```python
japan_mb = df[(df['parsed_place'].str.contains("Japan", case= False, na=False)) & (df['magType'] == 'mb')]

percentile_95 = np.percentile(japan_mb['mag'], 95)
```

```python
print("95Th percentile magnitude in Japan (mb):", percentile_95)
```

#### 2. Tsunami Percentage (Indonesia): 
Calculate the percentage of earthquakes in Indonesia that were associated with a tsunami.

```python
indonesia_eq = df[df['parsed_place'].str.contains("Indonesia", case= False, na= False)]

tsunami_count = indonesia_eq['tsunami'].sum()
total_eq = len(indonesia_eq)

tsunami_percentage = (tsunami_count/total_eq) * 100
```

```python
print("Tsunami percentage in Indonesia:", tsunami_percentage)
```

#### 4. Ring of Fire Classification:
Add a new column to the DataFrame called ring_of_fire that indicates whether the earthquake occurred in a country or U.S. state located on the
Ring of Fire.

Use the following locations:

Bolivia, Chile, Ecuador, Peru, Costa Rica, Guatemala, Mexico, *(Be careful not to include New Mexico)* Japan, Philippines, Indonesia, New Zealand, Antarctic
Canada, Fiji, Alaska, Washington, California
Russia, Taiwan, Tonga, Kermadec Islands

The column should contain:

● True if the earthquake occurred in one of these locations

● False otherwise

```python
ring_locations = {"Bolivia", "Chile", "Ecuador", "Peru", "Costa Rica", "Guatemala", "Mexico", "Japan", "Philippines", "Indonesia", "New Zealand", "Antarctic", "Canada", "Fiji", "Alaska", "Washington", "California", "Russia", "Taiwan", "Tonga", "Kermadec Islands"}
```

```python
def check_ring_of_fire(parsed_place):
    if pd.isna(parsed_place):
        return False
    
    for loc in ring_locations:
        if loc in parsed_place:
            if "New Mexico" in parsed_place:
                return False
            return True
    
    return False
```

```python
df['ring_of_fire'] = df['parsed_place'].apply(check_ring_of_fire)

print(df[['parsed_place', 'ring_of_fire']].tail())
```

#### 5. Earthquake Counts (Inside vs Outside Ring of Fire)
Calculate:

● The number of earthquakes that occurred within the Ring of Fire

● The number of earthquakes that occurred outside the Ring of Fire

```python
ring_counts = df['ring_of_fire'].value_counts()

inside_ring = ring_counts.get(True, 0)
outside_ring = ring_counts.get(False, 0)
```

```python
print("Earthquake occured within the Ring of fire:", inside_ring)
print("Earthquake occured outside the Ring of fire:", outside_ring)
```

#### 6. Tsunami Count Along the Ring of Fire
Find the total number of earthquakes in Ring of Fire locations that were associated with a
tsunami.

```python
ring_tsunami = df[(df['ring_of_fire'] == True) & (df['tsunami']==1)]

tsunami_ring_counts = len(ring_tsunami)
```

```python
print("Tsunami earthquakes in Ring of fire:", tsunami_ring_counts)
```
