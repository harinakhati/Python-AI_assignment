#### 4. Ring of Fire Classification:
Add a new column to the DataFrame called ring_of_fire that indicates whether the earthquake occurred in a country or U.S. state located on the
Ring of Fire.

Use the following locations:

Bolivia, Chile, Ecuador, Peru, Costa Rica, Guatemala, Mexico, *(Be careful not to include New Mexico)* Japan, Philippines, Indonesia, New Zealand, Antarctic
Canada, Fiji, Alaska, Washington, California
Russia, Taiwan, Tonga, Kermadec Islands

The column should contain:

● True if the earthquake occurred in one of these locations

● False otherwise

```python
ring_locations = {"Bolivia", "Chile", "Ecuador", "Peru", "Costa Rica", "Guatemala", "Mexico", "Japan", "Philippines", "Indonesia", "New Zealand", "Antarctic", "Canada", "Fiji", "Alaska", "Washington", "California", "Russia", "Taiwan", "Tonga", "Kermadec Islands"}
```

```python
def check_ring_of_fire(parsed_place):
    if pd.isna(parsed_place):
        return False
    
    for loc in ring_locations:
        if loc in parsed_place:
            if "New Mexico" in parsed_place:
                return False
            return True
    
    return False
```

```python
df['ring_of_fire'] = df['parsed_place'].apply(check_ring_of_fire)

df[['parsed_place', 'ring_of_fire']].tail()
```

#### 1. Filtering (earthquakes.csv):
Select all earthquakes that occurred in Japan with magType equal to 'mb' and magnitude ≥ 4.9.

```python
japan_quakes = earthquakes[(earthquakes['parsed_place'].str.contains("Japan", case = False, na = False)) & (earthquakes['magType']== 'mb') & (earthquakes['mag'] >= 4.9)]

japan_quakes
```

#### 2. Binning + Counts (earthquakes.csv): 
For earthquakes with magType equal to 'ml', create magnitude bins for each whole number (0–1, 1–2, 2–3, ...) and count how many earthquakes fall into each bin.

```python
ml_quakes = earthquakes[earthquakes['magType'] == 'ml'].copy()

bins = np.arange(0, ml_quakes['mag'].max() +1, 1)

ml_quakes['mag_bin'] = pd.cut(ml_quakes['mag'], bins)

bin_counts = ml_quakes['mag_bin'].value_counts().sort_index()

bin_counts
```

#### 3. Groupby + Resample + Aggregation (faang.csv): 
Group by ticker, resample to monthly frequency, and compute:

1. mean of open

2. max of high

3. min of low

4. mean of close

5. sum of volume

```python
faang['date'] = pd.to_datetime(faang['date'])

faang = faang.set_index('date').copy()

monthly_stats = (faang.groupby('ticker').resample('ME').agg({
    'open': 'mean',
    'high': 'max',
    'low': 'min',
    'close': 'mean',
    'volume': 'sum'
}))
```

```python
monthly_stats
```

#### 4. Crosstab with Aggregation (earthquakes.csv): 
Build a crosstab between tsunami (rows) and magType (columns). Instead of counts, show the maximum magnitude observed for each combination.

```python
cross_table = pd.crosstab(earthquakes['tsunami'], earthquakes['magType'], values = earthquakes['mag'], aggfunc = 'max')
```

#### 5. Z-scores with apply (faang.csv): 
For Netflix only (ticker == 'NFLX'), calculate the Z-score for each numeric column using apply().

```python
nflx = faang[faang['ticker'] == 'NFLX']

numeric_cols = nflx.select_dtypes(include = np.number)

z_scores = numeric_cols.apply(lambda x: (x-x.mean())/x.std())
```

#### 7. Transform to Index Values (faang.csv): 
Use transform() so that, for each ticker, all numeric values are expressed relative to the first date in that ticker’s data (divide each value by the ticker’s first-date value).

```python
faang = pd.read_csv('assign3_data/faang.csv')

faang['date'] = pd.to_datetime(faang['date'])

faang = faang.sort_values(['ticker', 'date']).copy()

numeric_cols = ['open', 'high', 'low', 'close', 'volume']

faang[numeric_cols] = faang.groupby('ticker')[numeric_cols].transform(lambda x: x / x.iloc[0])
```
